```python
import os

data_path = "data/pmc"
pdf_files = [x for x in os.listdir(data_path)]

print(len(pdf_files))
print(pdf_files[:3])
```

```python
import textract
from tqdm.auto import tqdm

parsing_failed_files = []
for pdf_file in (pbar := tqdm(pdf_files)):
    pbar.set_description(f"failed count : {len(parsing_failed_files)}")
    pdf_path = os.path.join(data_path, pdf_file)    
    try: 
        text = textract.process(pdf_path, method='pdfminer')
    except:
        parsing_failed_files.append(pdf_file)
        
```
